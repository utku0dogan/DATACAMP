**not randn ortalaması 0 standart sapması 1 olacak sekilde random sayı olusturur**

```python
mu1=100 #ortalama
sigma1=15 #standart sapma
x = mu1 + (sigma1*np.random.randn(10000))

mu2=110
sigma2= 10
y = mu2 + (sigma2*np.random.randn(10000))

#söz konusu verilerin histogramı

n, bins, patches = plt.hist(x,100,density=1,facecolor = 'b',alpha =0.25) #x', 100 adet bean olusturarak ciz
n, bins, patches = plt.hist(y,100,density=1,facecolor = 'g',alpha =0.25)

plt.xlabel("Veriler")
plt.ylabel("Olasılıklar")
plt.title(r"Verilerin Histogramı: $\mu_1=100, \ \mu_2 =110$") #yazima dikkat
plt.annotate('Gauss eğrisinin tepesi', xy=(110,0.04), xytext = (130,0.045000),
            arrowprops = dict(facecolor='black', shrink = 0.1))


plt.text(75, 0.025, r'$\mu_1 =100, \ \sigma_1=15$')
plt.text(120, 0.035, r'$\mu_2 =110, \ \sigma_2=10$')
plt.axis([40,160,0,0.05])
plt.grid(True)
plt.show()
```

# Seaborn Alıştırmalar

* Seaborn, Matplotlib kütüphanesi tabanlı, istatistiksel bir Pyhton veri görselleştirme kütüphanesidir.
* Seaborn kullanıcılara istatistiksel görselleştirme yapmaları için yüksek seviye bir arayüz sunar.
* Seaborn Kütüphanesinin işlevleri aşağıdaki gibidir. 
    * birden çok değişken arasındaki ilişkileri detaylı bir şekilde incelemek adnıa sunulan veri kümesi tabanlı bir uygulama programlama arayüzüdür.
    * Kategorik değişkenleri gözlemlemek ve üzerinde istatistiksel işlemler yapmak için destek sağlar.
    * Karmaşık görselleştirmeler oluşturmayı sağlayan çoklu grafikleri destekler.
    * Alt kümeler arasında karşılaştırma yapmak için tek değişkenli ve çift değişkenli görselleştirmeler sunar
    * Görselleştirmeyi daha anlaşılır hale getiren geniş renk skalası mevcuttur.
* Seaborn ve Matplotlib kütüphanesi arasındaki farklar:
    * Matplotlib kişiselleştirebilir.i fakat plot'ları daha anlaşılır, göze hitap eden şekle getirmek bir hayli zordur.
    * Seaborn sahip olduğu renk paleti, arayüzler ve grafik çeşitleri ile kullanıcıya farklı çözümler sunar.
    * Pandas Kütüphanesi ile çalışırken Seaborn kolaylık sağlar.

# Plotly Kütüphanesi

## Dünya Üniversite Sıralaması Veri Seti 
* Bu bölümde dünyada bulunan üniversitelerin farklı kriterlere göre sıralamasını içeren bir veri seti üzerine çalışacağız.
* Veri seri içerisinde bulunan sütunların isimleri ve açıklamaları
    1. world_rank - dünya sıralaması
    2. university_name - üniversite ismi
    3. country - üniversitenin bulunduğu ülke
    4. teaching - öğretim puanı
    5. international - uluslararası öğrenci/ çalışan açısından üniversite puanı
    6. research - araştırma puanı
    7. citations - alıntı puanı
    8. income gelir puanı
    9. total_score - toplam puan
    10. num_students - öğrenci sayısı
    11. student_staff_ratio - öğrenci/çalışan oranı
    12. international_students - uluslararası öğrenci oranı
    13. female_male_ratio - kadın/erkek öğrenci oranı
    14. year - sıralamanın yapıldığı yıl

```python
import plotly as py
from plotly.offline import init_notebook_mode, iplot, plot
init_notebook_mode(connected= True)
import plotly.graph_objs as go
import plotly.io as pio
import plotly.figure_factory as ff
```

```python
veri = pd.read_csv("University_Rankings_20220204.csv")
```

```python
veri.head()
```

```python
veri.columns
```

```python
veri.rename(columns= {'world_rank' : 'dunya_siralama',
                     'university_name' : 'uni_isim',
                     'country': 'ulke',
                     'teaching':'ogretim',
                     'international':'uluslararasi',
                     'research':'arastirma',
                     'citations':'alinti',
                     'income':'gelir',
                     'total_score':'toplam_puan',
                     'num_students' : 'ogrenci_sayisi',
                      'student_staff_ratio': 'ogrenci_calisan_orani',
                      'international_students':'uluslararasi_ogrenci',
                      'female_male_ratio':'kadin_erkek_orani',
                      'year':'yil'
                      
                      
                      
                     }, inplace = True)
```
